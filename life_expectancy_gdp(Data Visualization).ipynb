## Bar Charts To Compare Average
A) Create a bar chart from the data in df using Country on the x-axis and GDP on the y-axis.

```python
sns.barplot(x='Country', y='GDP', data=df)
plt.xticks(rotation=90)
plt.title('Comparing GDP')
plt.show()
plt.clf()
```

B) Create a bar chart using the data in df with Country on the x-axis and LEABY on the y-axis.

```python
sns.barplot(x='Country', y='LEABY', data=df)
plt.xticks(rotation=90)
plt.title('Comparing Life Expectancy at Birth')
plt.show()
plt.clf()
```

Both charts use the same six countries and the same bar‑chart style, so they look visually consistent. But they highlight very different patterns. The GDP chart shows huge economic differences between countries, with the United States and China far ahead. The life expectancy chart, on the other hand, has a much tighter range — most countries sit relatively close together, except for Zimbabwe, which stands out with a much lower value.

## Violin Plots To Compare Life Expectancy Distributions\

Next, I will create violin plots to compare the distribution of life expectancy values across all six countries.  
This visual will show the spread, density, and overall shape of life expectancy for each country, making it easier to compare their health outcomes.


```python
sns.violinplot(x='Country',y='LEABY', data=df, hue='Country')
plt.xticks(rotation=90)
plt.title('Comparing LEABY using a Violin plot')
plt.show()
plt.clf()
```

The violin plots show how life expectancy values are distributed for each country.  
Most countries have tall, narrow shapes, which indicates that their life expectancy has remained stable over the years with very little variation.

Zimbabwe is the clear exception.  
Its violin is wider and positioned lower on the scale, showing:
- a lower overall life expectancy  
- much more variation across the years  

This makes Zimbabwe stand out immediately from the other countries.

## Bar Plots Of GDP and Life Expectancy over time
Next, I will create bar plots to show how GDP and life expectancy change over time for each country.  
These visuals make it easy to compare year‑to‑year patterns and highlight how trends differ across countries.


```python
ax=sns.barplot(x='Country', y='GDP', hue='Year', palette='Set2', data=df)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1,1),loc='upper left')
plt.title('Comparing GDP over the past years')
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.show()
plt.clf()
```

```python
ax=sns.barplot(x='Country', y='LEABY', hue='Year', palette='Accent', data=df)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1,1))
plt.ylabel('Life expectancy at birth in years')
plt.title('Comparing LEABY over the past years')
plt.show()
```

## Scatter Plots of GDP and Life Expectancy Data

To compare GDP and life expectancy more effectively, I will create a series of scatter plots arranged on a shared figure.  
Using Seaborn’s FacetGrid makes it possible to generate multiple subplots at once, with each subplot representing a different year in the dataset.

In this visualization:
- GDP will be on the x‑axis  
- Life expectancy will be on the y‑axis  
- Each subplot will represent a single year  
- Countries will be differentiated by color  

This layout makes it easier to observe how the relationship between GDP and life expectancy changes over time and how each country compares within each year.

```python
g=sns.FacetGrid(df,col='Year', hue='Country', col_wrap=4,height=2)
g= (g.map(plt.scatter,'LEABY','GDP', edgecolor='w').add_legend())
plt.show()
plt.clf()
```

## Interpreting the FacetGrid Scatter Plots

Across the years, China shows the largest movement along the x‑axis.  
Its life expectancy increases steadily and noticeably from 2000 to 2015, creating a clear horizontal shift in the scatter points.

The United States shows the largest movement along the y‑axis.  
Its GDP rises sharply over the same period, producing a strong upward shift in the vertical direction.

These patterns are not surprising.  
China has experienced rapid improvements in public health and living conditions, while the United States has maintained one of the highest GDP levels globally with consistent growth.

Although the FacetGrid layout allows year‑by‑year comparison, the scatter plots can feel dense and visually busy.  
A single combined plot or a line‑based visualization may make the trends easier to follow.

## Line Plots for Life Expectancy

```python
g2=sns.FacetGrid(df,col='Country',col_wrap=2,height=4)
g2=(g2.map(sns.lineplot,'Year','LEABY').add_legend())
g2.fig.subplots_adjust(wspace=0.1, hspace=0.5)
g2.fig.set_size_inches(10,12)
plt.show()
plt.clf()
```

This grid shows how life expectancy changes over time for each country. Because every panel shares the same axes, it’s easy to compare the shape and steepness of the trends.

All countries span the same years, so the horizontal movement is identical. The vertical movement, however, varies: China shows the strongest upward shift in life expectancy, while South Africa moves noticeably but in a different pattern. Countries like the United States and Germany change more gradually.

The small‑multiple layout makes the overall patterns clear, but the panels are compact, and comparing exact values across countries can be difficult. A combined multi‑line plot offers a clearer way to compare countries directly on the same axes.


## Line Plots for Life GDP

```python
g3 = sns.FacetGrid(df, col='Country', col_wrap=2, height=4)
g3 = (g3.map(sns.lineplot,'Year','GDP').add_legend())
g3.fig.subplots_adjust(wspace=0.2)
g3.fig.set_size_inches(10,15)
plt.show()
plt.clf()
```

This grid shows how GDP changes over time for each country. Because all panels share the same axes, the differences in economic growth become easy to compare.

The strongest movement appears in China and the United States, both showing clear upward trends. Germany and Chile grow more steadily, while Mexico shows moderate variation. Zimbabwe stands out with much lower GDP values and a flatter trajectory.

The small‑multiple layout highlights each country’s pattern, but the compact size makes it harder to compare exact values across panels. A combined multi‑line plot offers a clearer view of how these economies diverge over time.
